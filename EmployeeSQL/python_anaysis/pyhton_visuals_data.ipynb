```python
import pandas as pd


# Read employees data
employees = pd.read_csv('employees.csv')

# Read salaries data
salaries = pd.read_csv('salaries.csv')

# Read departments data
departments = pd.read_csv('departments.csv')

# Display a few rows of each DataFrame to verify the data
print("Employees Data:")
employees.head()
```

## Average Salary by Gender:

> Analysis: Calculate the average salary for each gender and visualize the results.
> Visualization: A bar chart or box plot comparing the average salaries for male and female employees.


```python
import matplotlib.pyplot as plt
import seaborn as sns
merged_df = pd.merge(employees,salaries,on='emp_no', how='inner')
merged_df
```

```python
average_salary_by_gender = merged_df.groupby('sex')['salary'].mean()

# Visualization
sns.barplot(x=average_salary_by_gender.index, y=average_salary_by_gender.values)
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

## Employee Distribution by Department:

> Analysis: Explore the distribution of employees across different departments.
> Visualization: A pie chart or bar chart showing the percentage or count of employees in each department.

```python
 #Visualization

department_distribution = departments['dept_name'].value_counts()   
plt.figure(figsize=(8, 6))
department_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Employee Distribution by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
